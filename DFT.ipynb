Define function to calculate DFT

The DFT is defined as
$$
    X_k = \sum_{n=0}^{N-1} x_{n}\cdot e^{-i\tau k \frac{n}{N}}
$$
Where k is an input frequency such that:
$$
    \begin{align}
        k \in \mathbb{Z} \;\;\;\; k \in [0, N - 1]
    \end{align}
$$
and x is a list of floats that is 0-indexed

```julia
ðœ = 2 * pi

function DFT(data::Vector{Float64}, k::Number)
    N = length(data)
    
    result::Complex = 0
    for n in 0:N - 1
        result += data[n + 1] * exp(-im * ðœ * k * n / N) # Julia vectors are one-indexed hence [n + 1]
    end
    
    result
end
```

```julia
res = DFT(y, 2pi)

println(res)
println(abs(res))
println(angle(res))
        
scatter([real(res)/length(y)], [imag(res)/length(y)], color="blue", markersize=10)
```

```julia
dft_output = abs(res)/N * @. cos(2pi * x + angle(res))
```

```julia
plot(x, dft_output)
```
